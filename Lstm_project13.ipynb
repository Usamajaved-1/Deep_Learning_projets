```python
# now we assign the index to our every word
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([faqs])
tokenizer.word_index
```

```python
# now we split the sentences on the basis of line change
for sentence in faqs.split("/n"):
  print(sentence)
```

```python
# now we give number according to the sentences
for sentence in faqs.split('\n'):
  print(tokenizer.texts_to_sequences([sentence])[0])
  # by 0 all senences comes in single list
```

```python
# now we create a shape as explain on reg 205
input_text = [] # create a list to store it
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_text.append(tokenized_sentence[:i+1]) # string slicing
```

```python
input_text
```

```python
# now the size of all the rows are not equal so we apply zero padding
```

```python
# now we find out which sentence contain max letters
max_length = max([len(x) for x in input_text])
max_length
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_text,maxlen = max_length,padding = 'pre')
```

```python
padded_input_sequences
```

```python
# now we create input and output so last word is our output remainning is input
x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
x
```

```python
y
```

```python
x.shape,y.shape # y is scalar so we have to apply one hot encoding
```

```python
# so which type of problem is this ?
# it is a classification problem, bcz if we use regression values are in [0.2 0.3 0.4] but regresion is for 0 or 1
# in classification we use multi class we transfor our output with one hot encoding see pg 205
# in one hot encoding number are created according to unique numbers in sentence
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
len(tokenizer.word_index)
```

```python
y = to_categorical(y,num_classes =283) # total number of unique words in entence
# WE ENTERN 283 BCZ one hot encoding starts from 1 not 0

```

```python
y
```

```python
y.shape # every output is 283 num long sequence
```

```python
#  now we have x and y both
# now we create an architecture of our m
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(283,100,input_length = 56))# 283 total vocablury counts unique words  100 is embedding shape 56 bcz every row contain 56 words
model.add(LSTM(150))# 150 IS RANDOM
model.add(Dense(283,activation = 'softmax')) # mean total unique words and which words probability is highest that is the output
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs = 100)
```

```python
model.save('trained_lstm_model.keras')
```

```python
import numpy as np
#  now we predict some output from it
text =  'mail'
# tokenize
tokenized_text = tokenizer.texts_to_sequences([text])[0]
# padding
paded_input= pad_sequences([tokenized_text],maxlen =56,padding='pre')
#now prediction
# np.argmax(model.predict(paded_input)) # gives 283 dim vector give probality of all the words we have to find which word has highest probability arg max tell us at which position is the highest propbability 134
pos = np.argmax(model.predict(paded_input))
# we create a loop which number is at 134
for word,index in tokenizer.word_index.items():
  if index == pos:
    print(word)
```
