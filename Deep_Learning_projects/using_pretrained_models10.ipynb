```python
import tensorflow
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input,decode_predictions
from tensorflow.keras.preprocessing import image
```

```python
# we use keras in built pretraiend model here name resnet 50 also check register 177
model = ResNet50(weights="imagenet") # same weights we need as recieved on trainning image net database
```

```python
img_path = "/content/CHAIR.jpg"
img = image.load_img(img_path,target_size=(224,224))
x = image.img_to_array(img)
import numpy as np
x = np.expand_dims(x,axis=0) # convert image to batches
x = preprocess_input(x)

# explained from chatgpt

# Define the path to the image
# img_path = "/content/CHAIR.jpg"

# # Load the image and resize it to the required input size for the model (224x224)
# # target_size=(224, 224) ensures the image fits the model's input dimensions
# img = image.load_img(img_path, target_size=(224, 224))

# # Convert the image to a NumPy array (with pixel values)
# # This will produce an array of shape (224, 224, 3), representing height, width, and RGB channels
# x = image.img_to_array(img)

# # Import NumPy for array manipulations
# import numpy as np

# # Add an additional dimension to the image array to represent a batch
# # Converts the shape from (224, 224, 3) to (1, 224, 224, 3), where "1" represents a batch of size 1
# x = np.expand_dims(x, axis=0)

# # Preprocess the image to make it compatible with the input requirements of the pre-trained model
# # This step performs operations like mean subtraction or scaling, specific to the model being used
# x = preprocess_input(x)
```

```python
pred = model.predict(x)
print("predicted:" ,decode_predictions(pred,top = 3)[0])
```
