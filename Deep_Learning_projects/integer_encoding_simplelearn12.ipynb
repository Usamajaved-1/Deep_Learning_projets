```python
import numpy as np

docs = ['go pakistan',
		'pakistan pakistan',
		'hip hip hurray',
		'jeetega bhai jeetega pakistan jeetega',
		'Dil Dil Pakistan',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'Imran khan zindabad',
		'inquilab zindabad']
```

```python
# now we tokenize our whole content mean word by word sep capital to small remove special symbols
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token="<nothing>") # out of vocablury mean if new word come that is not in our dict replace it with nothing
```

```python
tokenizer.fit_on_texts(docs) # reg 189
```

```python
tokenizer.word_index #we can see which number is assign
```

```python
tokenizer.word_counts # which word come how many time
```

```python
tokenizer.document_count # how many total sentences
```

```python
# now the numbers assigned are in indexing shape
sequence = tokenizer.texts_to_sequences(docs)
```

```python
sequence # 10 for go 2 for pakistan
```

```python
# now we apply padding as on reg
from keras.utils import pad_sequences
sequence = pad_sequences(sequence,padding='post') # post mean add 0 at the end
```

```python
sequence # now the shape is ready
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import *
```

```python
(x_train,y_train),(x_test,y_test) = imdb.load_data()
```

```python
x_train[0] # if we se the shape it is alrwady in vectors
```

```python
# now we make all the reviews length equal
from tensorflow.keras.preprocessing.sequence import pad_sequences
x_train = pad_sequences(x_train,maxlen=50) #mean max legth of sentence is 50
x_test = pad_sequences(x_test,maxlen=50)  # add pooling
```

```python
x_train.shape,len(x_test[0]) #now  total length is 50 of all sentences
```

```python
x_train[0]
```

```python
from keras.layers import Dense,SimpleRNN,Flatten # see page no 190
model = Sequential()
model.add(SimpleRNN(32,input_shape = (50,1),return_sequences=False)) # shape mean 50 timpe step 1 input feature
# return sewquence mean if we want output at every time step outside if we false it mean it become inside the model and come when final output comes
model.add(Dense(1,activation="sigmoid"))
model.summary() # structure explain on reg 190 page no
```

```python
model.compile(loss= "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```python
model.fit(x_train,y_train,epochs = 5,validation_data= (x_test,y_test))
```
