```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D# as for ann dense layer for cnn cpn2d layer
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu', input_shape=(28,28,1)))
# conv2d explan above 32 are filters and every filter size is 3*3 padding explain pg#165 28*28*1 size of imge
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))

model.add(Flatten()) # model converted to 1d

model.add(Dense(128,activation='relu')) # tranfered to fully connected layer
model.add(Dense(10,activation='softmax'))
```

```python
model.summary() # first img is 28*28 then 26*26 and then 22*22 then converted to on 1 d n-f+1 on reg 28-3+1 = 26 and so on
```

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary() # in this padd is same mean model automatical apply padding on reg
```

how max_pooling is applied in keras

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu', input_shape=(28,28,1)))
# conv2d explan above 32 are filters and every filter size is 3*3 padding explain pg#165 28*28*1 size of imge
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding="valid")) # EXPLLALIN ON REG
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding="valid"))

model.add(Flatten()) # model converted to 1d

model.add(Dense(128,activation='relu')) # tranfered to fully connected layer
model.add(Dense(10,activation='softmax'))
```
