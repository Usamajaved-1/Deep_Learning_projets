```python
# pg186
from keras import Sequential
```

```python
from keras.layers import Dense,SimpleRNN
```

```python
model =Sequential()
model.add(SimpleRNN(3,input_shape = (4,5))) # 3 are unique features4,5 is the max shape on reg186
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
# 31 parametres total as explain in reg
print(model.get_weights()[0].shape)
```
